```python
import pandas as pd
import tabula as tb
```

```python
listapasta = !dir "./Notas Fiscais"
listapdfs = []
for i in range(len(listapasta[7:-2])):
    listapdfs.append(listapasta[7:-2][i][36:])
```

```python
dados = {'Data': [], 'Num.Nota': [], 'Unidade': [], 'Qtde': [], 'Produto': [], 
         'Razão Social e CNPJ Fornecedor': [], 'Endereço Fornecedor': [], 
         'Cidade, CEP e Fone Fornecedor': [], 'Razão Social e CNPJ Transportadora': [], 'Endereço Transportadora': []}

for n in listapdfs:
    arq = tb.read_pdf("./Notas Fiscais/" + n, pages="1")

    if len(arq) <= 5:
        for i in range(len(arq[len(arq)-1]['UNID'])):
            if (str(arq[len(arq)-1]['UNID'][i])).lower() != 'nan':
                try:
                    dados['Unidade'].append(str(arq[len(arq)-1]['UNID'][i]))
                except:
                    dados['Unidade'].append('Corrigir código')
                try:
                    dados['Qtde'].append(str(arq[len(arq)-1]['QUANT.'][i]).replace('.', ','))
                except:
                    dados['Qtde'].append('Corrigir código')
                try:
                    dados['Produto'].append(str(arq[len(arq)-1].iloc[i, 1]))
                except:
                    dados['Produto'].append('Corrigir código')
                try:
                    dados['Data'].append(arq[0].columns[0][(arq[0].columns[0].index('EMISSÃO:')+9):(arq[0].columns[0].
                                                                                            index('EMISSÃO:')+19)])
                except:
                    dados['Data'].append('Corrigir código')
                try:
                    dados['Num.Nota'].append(arq[0].iloc[0,0][3:])
                except:
                    dados['Num.Nota'].append('Corrigir código')
                try:
                    dados['Razão Social e CNPJ Fornecedor'].append(
                arq[0].columns[0][13:(arq[0].columns[0].index('OS PRODUTOS')-1)].title() + ' - CNPJ: ' + 
                arq[0].iloc[len(arq[0])-1, 0][-18:])
                except:
                    dados['Razão Social e CNPJ Fornecedor'].append('Corrigir código')  
                try:
                    dados['Endereço Fornecedor'].append(arq[0].iloc[7, 0][:arq[0].iloc[7, 0].index('0 - ENTRADA')].title())
                except:
                    dados['Endereço Fornecedor'].append('Corrigir código')
                try:
                    dados['Cidade, CEP e Fone Fornecedor'].append(arq[0].iloc[10, 0].title() + ' - ' + 
                                                          arq[0].iloc[13, 0][:36].replace(' FOLHA 1 /  1', '').title())
                except:
                    dados['Cidade, CEP e Fone Fornecedor'].append('Corrigir código')
                try:
                    dados['Razão Social e CNPJ Transportadora'].append(arq[len(arq)-2].iloc[0, 0].title() + ' - CNPJ: ' + 
                                                               arq[len(arq)-2].iloc[0, len(arq[len(arq)-2].columns)-1])
                except:
                    dados['Razão Social e CNPJ Transportadora'].append('Corrigir código')
                try:
                    dados['Endereço Transportadora'].append(arq[len(arq)-2].iloc[2, 0].replace('-  CEP', '').title())
                except:
                    dados['Endereço Transportadora'].append('Corrigir código')
            else:
                pass
       
    else:
        dados['Data'].append('Corrigir código')
        dados['Num.Nota'].append('Corrigir código') 
        dados['Unidade'].append('Corrigir código') 
        dados['Qtde'].append('Corrigir código')
        dados['Produto'].append('Corrigir código')
        dados['Razão Social e CNPJ Fornecedor'].append('Corrigir código') 
        dados['Endereço Fornecedor'].append('Corrigir código')
        dados['Cidade, CEP e Fone Fornecedor'].append('Corrigir código')
        dados['Razão Social e CNPJ Transportadora'].append('Corrigir código')
        dados['Endereço Transportadora'].append('Corrigir código')
```

```python
df = pd.DataFrame.from_dict(dados)
df.to_excel(r'./dadosNF.xlsx')
```
