## **Análise de dados do preço de medicamentos no Brasil**

### Exploração via dataframe da biblioteca pandas

Fonte dos dados: https://dados.gov.br/dataset/ef3d5f8b-1661-4381-a369-4aed487bf313/resource/84531c98-b162-4c00-ae21-002192a51051

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import sqlite3
import os
```

```python
df = pd.read_csv(r'/home/mribeiro/Downloads/TA_PRECO_MEDICAMENTO.csv', low_memory = False)
print(list(df.columns[i] for i in range (len(list(df)))))
```

```python
lab10 = df.groupby(['LABORATÓRIO']).count()['SUBSTÂNCIA'].sort_values(ascending = False).\
head(10).reset_index().sort_values('SUBSTÂNCIA')
plt.figure(figsize = (12, 4))
plt.barh(lab10['LABORATÓRIO'], lab10['SUBSTÂNCIA'], label = list('SUBSTÂNCIA'), color = 'green')
plt.xticks([])
for i, v in enumerate(lab10['SUBSTÂNCIA']):
    plt.text(v + 10, i, str(v))
plt.title('Os 10 laboratórios com maior número de produtos registrados')
plt.show()

#ax = lab10.plot.barh(color = 'green')
#ax.set_title('Os 10 laboratórios com mais produtos registrados')
#ax.set_yticklabels(lab10['LABORATÓRIO'])
#ax.invert_yaxis()
#ax.get_xaxis().set_visible(False)
#plt.show()
```
